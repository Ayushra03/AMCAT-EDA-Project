```python
def str_corr(inputs):
   df[inputs] = df[inputs].str.lower().str.strip()
```

```python
Strings = ['Designation','JobCity','10board','12board','Specialization','CollegeState']
```

```python
def make2other(df, inputs):
    for Designation in df[inputs].unique():
        min_count = df[inputs].value_counts()[:10].min()
        if df[df[inputs] == Designation][inputs].value_counts()[0] < min_count:
            df.loc[df[inputs] == Designation, inputs] = 'other'
```

```python
for columns in Strings:
    make2other(df, columns)
```
